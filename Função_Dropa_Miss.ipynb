# Função Dropa Miss
Função que dropa colunas com uma porcentagem de valores miss
superior a um valor P_miss arbitrário, ou seja, P_miss é a menor % tolerada de miss.

**A função recebe** dois parâmetros: o dataframe **df** que contêm as colunas que serão analisadas pela função e o **P_miss**, que é o valor da menor % tolerada de miss na coluna (o valor default é igual a 0%).   
**A função devolve** o **df atualizado** (com as colunas com mais de P_miss% dropadas) e um **relatório** (na forma de dataframe) mostrando a porcentagem de miss de cada coluna (olhando esse relatório podemos ver quais colunas que foram dropadas, que são as que possuem porcentagem maior que P_miss%). 


```
def DropaMiss(df, P_miss=0):
  df_miss2 = ((df.isna().sum()/len(df))*100).to_frame().rename(columns={0:'P_miss'}).reset_index().sort_values('P_miss', ascending = False)
  df_miss = ((df.isna().sum()/len(df))*100).to_frame().rename(columns={0:'P_miss'}).reset_index()
  df_miss = df_miss[df_miss['P_miss'] <= P_miss]
  df = df[df_miss['index'].to_list()]
  return df,df_miss2
```

### Aplicação da função Dropa Miss

```
import pandas as pd
import seaborn as sns
```

```
titanic = sns.load_dataset("titanic")
df = titanic
df
```

```
#usando a função DropaMiss e gerando o relatorio 'relatorio_miss'.
#Vamos definir P_miss igual a 75%
#ou seja, vamos dropar, se existirem, colunas com mais de 75% de valores miss
df, relatorio_miss = DropaMiss(df, 75)
```
